```python
import tkinter as tk
import webbrowser
import os

def open_text_file():
    file_path = "likes.html"

    if not os.path.exists(file_path):
        with open(file_path, "w") as file:
            file.write("")

    webbrowser.open(file_path)

def delete_text_file():
    file_path = "likes.html"
    if os.path.exists(file_path):
        os.remove(file_path)
        info_label.config(text="File 'likes.html' deleted.")
    else:
        info_label.config(text="File 'likes.html' does not exist.")


root = tk.Tk()
root.title("Text File Editor")

open_button = tk.Button(root, text="Open 'likes.html' in Text Editor", command=open_text_file)
open_button.pack(pady=20)

delete_button = tk.Button(root, text="Delete 'likes.html' File", command=delete_text_file)
delete_button.pack(pady=10)

close_button = tk.Button(root, text="Close", command=lambda: root.destroy())
close_button.pack()


root.mainloop()
```

```python
from bs4 import BeautifulSoup   
```

```python
links = []
with open('likes.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
soup = BeautifulSoup(html_content, 'html.parser')
all_links = soup.find_all('a', class_='sc-link-primary')
for link in all_links:
    href = link['href']
    print(href)
    links.append(href)
```

```python
full_links = []
link_string = "https://soundcloud.com"
for link in links:
    full_link = link_string + link
    full_links.append(full_link)
print(full_links)
```
